# Control flow - Loops and Conditionals.

```
data = ['a', 'b', 'c', 'd', 'e']

for i in range(10):
  print(i, end=" ")

print(" ")
for letter in data:
  print(letter, end=" ")

for i, letter in enumerate(data):
  print(i, letter, end=" ")
```

```
data = ['a', 'b', 'c', 'd', 'e']

i = 0

while i < len(data):
  print(data[i], end=" ")
  i += 1

while True:
  x = input()
  if x == 'stop':
    break
  elif x == 'special':
    continue
  print("It wasn't stop or special")
  print("printing data")
```

```
data_dictionary = {
    'Yo': 5,
    'Tony': 15,
    'Jimmy': 25,
    'Sam': 12,
    'Monica': 30,
    'Yo': 10
}

print(data_dictionary)

for key, value in data_dictionary.items():
  if value == 12:
    print('We found 12')
    break
  print(key, value)
```

# Guessing game

- Generate a random number - random module.
- do while - Python doesn't have this built in.
- do something once - possibly repeat it.
- Ask us for a number if we get it wrong, ask again.

logic for right/wrong: guessed == correct value

```
from random import randint

low = 0
high = 100

correct = randint(0, 100)
print(f"Guess a number between {low} and {high}")

while True:
  guess = int(input())

  if guess < correct:
    print('Guess higher')
  elif guess > correct:
    print('Guess lower')
  else:
    print('You win')
    break
```
