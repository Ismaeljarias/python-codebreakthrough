# When scrapping - Beautiful Soup
when scraping is having a computer work with web pages through their html, css, etc.

This is the only option for working with websites when there is not API or of the API foes not give the needed options.

```python
from IPython.display import Image, display

Image('https://images.woopicx.com/637ae945-3e7e-4b2c-b576-73d23cb56bfc_small.jpg')
```

```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Image, display
import mimetypes

response = requests.get('https://woopicx.com/')

soup = BeautifulSoup(response.text, 'html.parser')
for img in soup.find_all('img')[:4]:
  full_img = img['src']
  mimetype = mimetypes.guess_type(full_img)

  if mimetype[0] == 'image/jpeg':
    display(Image(full_img, width=100))
```

# Dates and Time

```python
from datetime import datetime

now = datetime.now()
print(now)

future = datetime(2026, 1, 1)
print(future)

time_dif = future - now
print(f'time until 2026: {time_dif.days}')
```
